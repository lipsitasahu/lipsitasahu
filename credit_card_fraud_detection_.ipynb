```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Only use the 'Amount' and 'V1', ..., 'V28' features 
features = ['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
        'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
        'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28','Amount']
# The target variable which we would like to predict, is the 'Class' variable 
target = 'Class'
# Now create an X variable (containing the features) and an y variable (containing only the target)
X = df[features]
y = df[target]
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()

X=scaler.fit_transform(X)
```

```python
sns.distplot(df.V1)
```

```python
col=df.columns
import matplotlib.pyplot as plt 
for i in col:
    sns.distplot(df[i])
    plt.show()
```

```python
#Split the data set using 'train_test_split' function 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Instantiate the model to an empty object 
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression()
model
```

```python
# Train the model using 'fit' method 
model.fit(X_train,y_train)
```

```python
# Test the model using 'predict' method 
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score 
print(100*accuracy_score(y_test, y_pred))
```
